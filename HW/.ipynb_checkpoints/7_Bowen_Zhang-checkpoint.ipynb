## B. Comparing moon diameters and volumes to their host planets


```python
import pandas as pd
import numpy as np
```

```python
moons = pd.read_excel('../data/moons.xlsx', index_col = None, header = 0)
planets = pd.read_excel('../data/planets.xlsx', index_col = None, header = 0)
```

```python
MP = pd.merge(planets,moons,  on='planet_name')
```

```python
from numpy import pi
import matplotlib.pyplot as plt
```

```python
MP['moon_volume_km3'] = [4/3 * pi * (dia/2)**3 for dia in MP['moon_diameter_km'] ]
MP['planet_volume_km3'] = [4/3 * pi * (dia/2)**3 for dia in MP['planet_diameter_km'] ]
```

```python
fig,ax = plt.subplots(figsize=(8,8))
ax.set_xlabel("planet_diameter_km")
ax.set_ylabel("moon_diameter_km")
for i in range (MP.index.shape[0]):
    ax.scatter(MP['planet_diameter_km'][i],MP['moon_diameter_km'][i], s = MP['moon_diameter_km'][i] )
    ax.annotate(MP['moon_name'][i], color='k', 
                xy=(MP.planet_diameter_km[i], MP.moon_diameter_km[i]))


fig.savefig('Diameter.pdf')
```

```python
fig,ax = plt.subplots(figsize=(8,15))
ax.set_xlabel("planet_volume_km3")
ax.set_ylabel("moon_volume_km3")
for i in range (MP.index.shape[0]):
    ax.scatter(MP['planet_volume_km3'][i],MP['moon_volume_km3'][i], s = MP['moon_diameter_km'][i] )
    ax.annotate(MP['moon_name'][i], color='k', 
                xy=(MP.planet_volume_km3[i], MP.moon_volume_km3[i]))

fig.savefig('Volume.pdf')
```

## C. Comparing moon and sun angular diameters

```python
MP['moon_angular_diameter_arcsec'] = [206265 * Dia / Dis  for Dia,Dis  in zip(MP['moon_diameter_km'] ,MP['moon_planet_distance_km'])]
MP['sun_angular_diameter_arcsec'] = [206265 * Dia / Dis   for Dia,Dis  in zip(MP['planet_diameter_km'] ,MP['planet_sun_distance_km'])]
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(10,10))
ax.set_xlabel("sun_angular_diameter_arcsec")
ax.set_ylabel("moon_angular_diameter_arcsec")
#ax.scatter(MP['sun_angular_diameter_arcsec'],MP['moon_angular_diameter_arcsec'], s=MP['moon_diameter_km'])
for i in range (MP.index.shape[0]):
    ax.scatter(MP['sun_angular_diameter_arcsec'][i],MP['moon_angular_diameter_arcsec'][i], s = MP['moon_diameter_km'][i])
    ax.annotate(MP['moon_name'][i], color='k', 
                xy=(MP.sun_angular_diameter_arcsec[i], MP.moon_angular_diameter_arcsec[i]))
x = np.linspace(0,np.max(MP['sun_angular_diameter_arcsec']))
ax.plot(x,x, label="1:1 ratio")
ax.legend();
fig.savefig('angular_diameter_arcsec.pdf')
```
