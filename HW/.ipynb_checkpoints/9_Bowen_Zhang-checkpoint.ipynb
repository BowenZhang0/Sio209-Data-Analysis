```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

### A. Student survey data wrangling

```python
S_15 = pd.read_csv('../data/survey_scores_2015_T.csv')
```

```python
S_18 = pd.read_csv('../data/survey_scores_2018.csv')
```

```python
S_15.dtypes
```

```python
for i in S_15.index:
    if S_15.computer_has[i] == 'N':
        S_15.computer_has[i] = 'False' 
    else:
        S_15.computer_has[i] = 'True' 
```

```python
S_15
```

```python
S_18.drop(['Timestamp' ,'Graduate program and advisor'], axis=1, inplace=True)
```

```python
S_18.columns = [S_15.columns[0],S_15.columns[1],S_15.columns[2],S_15.columns[4],
                S_15.columns[5],S_15.columns[6],S_15.columns[7],S_15.columns[8]]
```

```python
for i in S_18.index:
    if S_18.computer_has[i] == 'N':
        S_18.computer_has[i] = 'False' 
    else:
        S_18.computer_has[i] = 'True' 
```

```python
Col_15 = S_15.columns
Col_18 = S_18.columns
```

```python
for j in range (5):
    for i in S_18.index:
        if S_18[Col_18[j+3]][i] == 'None':
            S_18[Col_18[j+3]][i] = 0
        elif S_18[Col_18[j+3]][i] == 'Some':
            S_18[Col_18[j+3]][i] = 1
        elif S_18[Col_18[j+3]][i] == 'Moderate':
            S_18[Col_18[j+3]][i] = 2
        elif S_18[Col_18[j+3]][i] == 'Experienced': 
            S_18[Col_18[j+3]][i] = 3
            
            
```

```python
S_18
```

### B. Statistical analysis

```python
S_15.computer_os.value_counts()
print('In 2015, the proportion of Mac and Windows users are %s and %s respectively'
      %(S_15.computer_os.value_counts()[0]/S_15.computer_os.value_counts().sum(),
       S_15.computer_os.value_counts()[1]/S_15.computer_os.value_counts().sum()))
```

```python
S_18.computer_os.value_counts()
print('In 2018, the proportion of Mac and Windows users are %s and %s respectively'
      %((S_18.computer_os.value_counts()[0]+S_18.computer_os.value_counts()[2])/S_18.computer_os.value_counts().sum(),
       (S_18.computer_os.value_counts()[1]+S_18.computer_os.value_counts()[3])/S_18.computer_os.value_counts().sum()))
```

```python
f = pd.DataFrame([[S_15.computer_os.value_counts()[0], S_15.computer_os.value_counts()[1]],
                  [S_18.computer_os.value_counts()[0]+S_18.computer_os.value_counts()[2],
                   S_18.computer_os.value_counts()[1]+S_18.computer_os.value_counts()[3]]], 
                 index=['2015', '2018'], 
                 columns=['Mac', 'Window'])
f
```

```python
odds_ratio, p_value = stats.fisher_exact(f, alternative='two-sided')
odds_ratio, p_value
```

```python
for i in range(1,6):
    t, p = stats.ttest_ind(S_15[Col_15[-i]], S_18[Col_15[-i]], equal_var=False)
    print("for %s     ttest_ind: t = %s, p = %s" % (Col_15[-i],t, p))
    if p >= 0.05:
        print('the two samples have equal means\n')
    else:
        print('the two samples\' means are not equal\n')
```

### C. Stock data

```python
import pandas_datareader as pdr
start_pdr = '2015/12/3'
end_pdr = '2018/12/3'
A = pdr.data.DataReader('A', 'iex', start_pdr, end_pdr)
A.set_index(pd.to_datetime(A.index), inplace=True)

SPY = pdr.data.DataReader('SPY', 'iex', start_pdr, end_pdr)
SPY.set_index(pd.to_datetime(SPY.index), inplace=True)
```

```python
A['year'] = [A.index[i].year for i in range (len( A.index)) ]
SPY['year'] = [SPY.index[i].year for i in range (len(SPY.index)) ]
```

```python
print('Agilent Tech Inc.')
print('The hightest price in the 2018 is %s'%A[A.year == 2018]['high'].max() )
print('The lowest price in the 2018 is %s\n'%A[A.year == 2018]['low'].min() )
print('S&P 500')
print('The hightest price in the 2018 is %s'%SPY[SPY.year == 2018]['high'].max() )
print('The lowest price in the 2018 is %s\n'%SPY[SPY.year == 2018]['low'].min() )
```

```python
fig,ax = plt.subplots()
ax.plot(A.index,A.close, label = 'Agilent Tech Inc.')
ax.set_ylabel('close_over',fontsize = 20 ,color = 'blue')
for label in ax.get_yticklabels():
    label.set_color('blue')
ax.legend()

ax_2 = ax.twinx()
ax_2.plot(SPY.index,SPY.close,label = 'S&P 500',color = 'red')
for label in ax_2.get_yticklabels():
    label.set_color('red')
ax_2.legend(loc =4)
```

```python
print('For Agilent Tech Inc, the close price changed percent is %s\n'%((A.close[A.index[-1]]-A.close[0])/A.close[0]))
print('For S&P 500, the close price changed percent is %s'%((SPY.close[SPY.index[-1]]-SPY.close[0])/SPY.close[0]))
```
